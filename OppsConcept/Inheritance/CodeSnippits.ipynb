#### Call by Value , Not change in i , j value 

```
class Test {
  /* This method causes no change to the arguments
     used in the call. */
  public void NoChange(int i, int j) {
    i = i + j;
    j = -j;
  }
}
```

```
Test ob = new Test();

int a = 15, b = 20;
Console.WriteLine("a and b before call: " +
                a + " " + b);
ob.NoChange(a, b);
Console.WriteLine("a and b after call: " +
                a + " " + b);
```

### Call by Reference without the ref keyword

```
// Objects are passed by reference.
class Test {
  public int a, b;
  public Test(int i, int j) {
    a = i;
    b = j;
  }
  /* Pass an object. Now, ob.a and ob.b in object
     used in the call will be changed. */
  public void Change(Test ob) {
    ob.a = ob.a + ob.b;
    ob.b = -ob.b;
  }
}
```

```
Test ob = new Test(10,15);
ob.Change(ob);
Console.WriteLine("ob.a and ob.b before call: " + ob.a + " " + ob.b);
```

## Call by Reference with Ref keyword 

```
class RefTest
{
    // Call by reference
    public void Square(ref int number)
    {
        number = number*number;
    }
}
```

```
RefTest ob = new RefTest();
int a = 10;
Console.WriteLine("a before call: " + a);
ob.Square(ref a); // notice the use of ref
Console.WriteLine("a after call: " + a);
```

##  Call By Value

```
class RefTestByValue
{
    // Call by reference
    public void Square(int number)
    {
        number = number*number;
    }
}
```

```
RefTestByValue ob = new RefTestByValue();
int a = 10;
Console.WriteLine("a before call: " + a);
ob.Square(a); // notice the use of ref
Console.WriteLine("a after call: " + a);
```
